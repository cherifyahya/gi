```python
import pandas as pd

# 1. Charger le fichier CSV dans une variable appelée 'dataset'
dataset = pd.read_csv('Salary_dataset.csv')

# 2. Afficher les 5 premières lignes pour vérifier le chargement
dataset.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Séparer les données en X (features) et y (target)
# On utilise [[ ]] pour X car l'algorithme attend une matrice 2D
X = dataset[['YearsExperience']].values
y = dataset['Salary'].values

# 2. Créer et entraîner le modèle avec la bibliothèque
modele = LinearRegression()
modele.fit(X, y)

# 3. Récupérer et afficher les paramètres mathématiques calculés
theta_1 = modele.coef_[0]
theta_0 = modele.intercept_
print(f"Pente calculée (theta_1) : {theta_1:.4f}")
print(f"Ordonnée à l'origine (theta_0) : {theta_0:.4f}")

# 4. Visualiser le résultat
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', alpha=0.6, label='Données réelles')
# J'utilise une ligne noire pour le modèle pour une lecture claire
plt.plot(X, modele.predict(X), color='black', linewidth=2, label='Modèle Scikit-Learn')

plt.title("Salaire vs Années d'expérience")
plt.xlabel("Années d'expérience")
plt.ylabel('Salaire')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
