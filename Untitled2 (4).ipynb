```python
import pandas as pd

# 1. Charger le fichier CSV dans une variable appelée 'dataset'
dataset = pd.read_csv('Salary_dataset.csv')

# 2. Afficher les 5 premières lignes pour vérifier le chargement
dataset.head()
```

```python
#----Régression linéaire simple----

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1. Séparer les données
X = dataset[['YearsExperience']].values
y = dataset['Salary'].values

# 2. Créer et entraîner le modèle
modele = LinearRegression()
modele.fit(X, y)

# 3. Prédictions
y_pred = modele.predict(X)

# 4. Afficher les résultats mathématiques
print("Pente:", modele.coef_[0])
print("Ordonnée:", modele.intercept_)
print("MSE :", mean_squared_error(y, y_pred))
print("Score R2 :", r2_score(y, y_pred))

# 5. Visualisation simple
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='black')
plt.title("Salaire selon l'expérience")
plt.xlabel("Années d'expérience")
plt.ylabel("Salaire")
plt.show()
```

```python
# Régression polynomiale (Degré 2)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score # Import des métriques

# 1. Données
dataset = pd.read_csv('Salary_dataset.csv')
X = dataset[['YearsExperience']].values
y = dataset['Salary'].values

# 2. Transformation Polynomiale (Degré 2)
transformeur_poly = PolynomialFeatures(degree=2)
X_poly = transformeur_poly.fit_transform(X)

# 3. Entraînement
modele = LinearRegression()
modele.fit(X_poly, y)

# Prédiction sur les vrais points pour calculer l'erreur
y_pred = modele.predict(X_poly)

print("Pentes (Coefficients) :", modele.coef_)
print("Ordonnée (Intercept)  :", modele.intercept_)
print("MSE (Erreur Moyenne)  :", mean_squared_error(y, y_pred))
print("Score R2 (Précision) :", r2_score(y, y_pred))

# 4. Visualisation
plt.scatter(X, y, color='blue', label='Vrais salaires')
X_lisse = np.arange(min(X), max(X), 0.1).reshape(-1, 1)
y_courbe = modele.predict(transformeur_poly.fit_transform(X_lisse))

plt.plot(X_lisse, y_courbe, color='black', label='Courbe polynomiale')
plt.title("Régression Polynomiale (Degré 2)")
plt.xlabel("Années d'expérience")
plt.ylabel("Salaire")
plt.legend()
plt.show()
```

```python
# Régression polynomiale (Degré 10)
#Avec un degré 10, le modèle va forcer la courbe à passer par chaque point

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score # Import des métriques

# 1. Données
dataset = pd.read_csv('Salary_dataset.csv')
X = dataset[['YearsExperience']].values
y = dataset['Salary'].values

# 2. Transformation Polynomiale (Degré 10)
transformeur_poly = PolynomialFeatures(degree=10)
X_poly = transformeur_poly.fit_transform(X)

# 3. Entraînement
modele = LinearRegression()
modele.fit(X_poly, y)

# Prédiction sur les vrais points pour calculer l'erreur
y_pred = modele.predict(X_poly)

print("Pentes (Coefficients) :", modele.coef_)
print("Ordonnée (Intercept)  :", modele.intercept_)
print("MSE (Erreur Moyenne)  :", mean_squared_error(y, y_pred))
print("Score R2 (Précision) :", r2_score(y, y_pred))

# 4. Visualisation
plt.scatter(X, y, color='blue', label='Vrais salaires')
X_lisse = np.arange(min(X), max(X), 0.1).reshape(-1, 1)
y_courbe = modele.predict(transformeur_poly.fit_transform(X_lisse))

plt.plot(X_lisse, y_courbe, color='black', label='Courbe polynomiale')
plt.title("Régression Polynomiale (Degré 10)")
plt.xlabel("Années d'expérience")
plt.ylabel("Salaire")
plt.legend()
plt.show()
```

```python
#------ Régression linéaire multiple-------
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1. Générer un dataset fictif
# n_features=3 veut dire qu'on a 3 variables d'entrée (X1, X2, X3)
X_multi, y_multi = make_regression(n_samples=100, n_features=3, noise=15, random_state=42)

# 2. Créer et entraîner le modèle
modele_multi = LinearRegression()
modele_multi.fit(X_multi, y_multi)

# 3. Prédictions
y_pred_multi = modele_multi.predict(X_multi)

# 4. Afficher les résultats mathématiques
print("Pentes (une par variable) :", modele_multi.coef_)
print("Ordonnée à l'origine      :", modele_multi.intercept_)
print("MSE                       :", mean_squared_error(y_multi, y_pred_multi))
print("Score R2                  :", r2_score(y_multi, y_pred_multi))

# 5. AJOUT DU GRAPHIQUE : Comparaison Réalité vs Prédiction
plt.figure(figsize=(10, 6))

# On trace les points (Vrai y en abscisse, y prédit en ordonnée)
plt.scatter(y_multi, y_pred_multi, color='blue', alpha=0.6, label='Données')

# On trace la ligne de perfection (y = x)
# Si le point est sur cette ligne, l'IA a deviné le prix exact.
plt.plot([y_multi.min(), y_multi.max()], [y_multi.min(), y_multi.max()],
         color='red', lw=2, linestyle='--', label='Modèle Parfait')

plt.title("Régression Multiple : Réalité vs Prédictions")
plt.xlabel("Valeurs Réelles")
plt.ylabel("Valeurs Prédites")
plt.legend()
plt.grid(True)
plt.show()
```
